```python
import os
import argparse
import sys
import platform
import time
import configparser as ConfigParser
from shutil import copyfile

from images import *
from metric import*
import tensorflow as tf
import pandas as pd
tf.keras.backend.set_image_data_format('channels_last')
import tensorflow.keras.backend as K

# Training settings
st = Settings()
options = st.get_options()

# set task to train
options['task'] = 'inference'
current_path = os.getcwd()
datasets_path = os.path.join('/'.join(current_path.split('/')[:current_path.split('/').index('Code')]), 'DataSets')
options['test_folder'] = os.path.join(datasets_path, options['test_folder'])
# list scans
scan_list = os.listdir(options['test_folder'])
scan_list.sort()
```

```python
accuracy_df=pd.DataFrame(columns=['patient_id','study','dice_coefficient','tpr','fpr'])
#with tf.compat.v1.Session() as sess:
for scan in scan_list:
    patient_id,study = scan.split('_')[0],scan.split('_')[1]
    path=os.path.join(options['test_folder'],scan)
    lesion_path = os.path.join(path, 'tmp/lesion.nii.gz')
    predicted_path = os.path.join(path, options['experiment'],options['experiment'] +'_hard_seg.nii.gz')
    predicted_path = os.path.join(path, options['experiment'],options['experiment'] +'_prob_1.nii.gz')
    
    target=load_img(lesion_path)
    predict=load_img(predicted_path)

    target=np.expand_dims(np.expand_dims(target,axis=0),-1)
    predict=np.expand_dims(np.expand_dims(predict,axis=0),axis=-1)
    dice=0
    with tf.compat.v1.Session() as sess:
        dice=dice_coefficient(target, predict)
    #cm = confusion_matrix(target, predict) # careful with the order of arguments!
    #tn, fp, fn, tp = cm.ravel()
    #TPR_value = tp/(tp+fn)
    #mse=tf.keras.metrics.MeanSquaredError()
    #pre=tf.keras.metrics.Precision()
    #recall=tf.keras.metrics.Recall()
    tp=tf.keras.metrics.TruePositives()
    fn=tf.keras.metrics.FalseNegatives()
    fp=tf.keras.metrics.FalsePositives()

    #mse.update_state(target,predict)
    #pre.update_state(target,predict)
    #recall.update_state(target,predict)
    tp.update_state(target,predict)
    fn.update_state(target, predict)
    fp.update_state(target, predict)

    #r1=mse.result().numpy()
    #r2=pre.result().numpy()
    #r3=recall.result().numpy()
    tp_value, fn_value, fp_value = 0.1,0.1,0.1
    #tp_value=tp.result().numpy()
    with tf.compat.v1.Session() as sess:
        pass
        #tp_value=(tp.result()).eval()
        #fn_value=fn.result().eval()
        #fp_value=fn.result().eval()

    tpr=tp_value/(tp_value+fn_value)
    fpr=fp_value/(fp_value+tp_value)
    #fnr=r5/(r4+r5)
    #print(r1,r2,r3,r4,r5,tpr,fnr)
    #tp.reset_state()
    #fn.reset_state()
    #tpr=TPR(np.expand_dims(np.expand_dims(target,axis=0),-1),np.expand_dims(np.expand_dims(predict,axis=0),axis=-1))

    df = pd.DataFrame([[patient_id,study,dice, tpr, fpr ]], columns=['patient_id','study','dice_coefficient','tpr','fpr'])
    accuracy_df=accuracy_df.append(df)

print(accuracy_df)
```
